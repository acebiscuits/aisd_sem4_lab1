```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
```

```python
def read_data(filename: str)->tuple:
    df = pd.read_csv(filename, header=None)
    points = df[0].to_string(index=False)
    v_time = df[1].to_string(index=False)
    tree_time = df[2].to_string(index=False)
    return points, v_time, tree_time

def graph_create(data_1: tuple, data_2: tuple, data_3: tuple)->None:
    fig, ax = plt.subplots(3, 2)
    fig. tight_layout (h_pad= 4, w_pad=3 )
    x = [float(x) for x in data_1[0].split()]
    y = [float(y) for y in data_1[1].split()]
    ax[0, 0].plot(x, y)
    ax[0, 0].set_title("std::vector: filling")
    ax[0, 0].set_xlabel('points')
    ax[0, 0].set_ylabel('time')
    y = [float(y) for y in data_1[2].split()]
    ax[0, 1].plot(x, y)
    ax[0, 1].set_title("tree: filling")
    ax[0, 1].set_xlabel('points')
    ax[0, 1].set_ylabel('time')
    y = [float(y) for y in data_2[1].split()]
    ax[1, 0].plot(x, y)
    ax[1, 0].set_title("std::vector: finding")
    ax[1, 0].set_xlabel('points')
    ax[1, 0].set_ylabel('time')
    y = [float(y) for y in data_2[2].split()]
    ax[1, 1].plot(x, y)
    ax[1, 1].set_title("tree: finding")
    ax[1, 1].set_xlabel('points')
    ax[1, 1].set_ylabel('time')
    y = [float(y) for y in data_3[1].split()]
    ax[2, 0].plot(x, y)
    ax[2, 0].set_title("std::vector: add and delete")
    ax[2, 0].set_xlabel('points')
    ax[2, 0].set_ylabel('time')
    y = [float(y) for y in data_3[2].split()]
    ax[2, 1].plot(x, y)
    ax[2, 1].set_title("tree: add and delete")
    ax[2, 1].set_xlabel('points')
    ax[2, 1].set_ylabel('time')
```

```python
filling_data = read_data('filling_speed.csv')
finding_data = read_data('finding_speed.csv')
adding_deleting_data = read_data('adding_deleting_speed.csv')
graph_create(filling_data, finding_data, adding_deleting_data)
```
